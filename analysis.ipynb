```python
def extract_feat(text):
    if isinstance(text, float) and np.isnan(text):
        return [np.nan] * 7  # Return a list of NaNs if input is NaN
    try:
        if float(text):
            return [text] + [''] * 6  # Return text as the first element
    except ValueError:
        pass
    return text.split(' ')[:7]  # Split text and take first 7 elements

# Extracting features from 'saledate' column
car_1['sales_day'] = car_1['saledate'].apply(lambda x: extract_feat(x)[0])
car_1['sales_month'] = car_1['saledate'].apply(lambda x: extract_feat(x)[1])
car_1['sales_day_date'] = car_1['saledate'].apply(lambda x: float(extract_feat(x)[2]) if extract_feat(x)[2] else np.nan)
car_1['sales_year'] = car_1['saledate'].apply(lambda x: float(extract_feat(x)[3]) if extract_feat(x)[3] else np.nan)
car_1['sales_time'] = car_1['saledate'].apply(lambda x: extract_feat(x)[4])
car_1['sales_GMT'] = car_1['saledate'].apply(lambda x: extract_feat(x)[5])
car_1['sales_Region'] = car_1['saledate'].apply(lambda x: extract_feat(x)[6])
```

```python
# checking for the percentage of outliers in the 'odometer' and 'mmr' columns
def percentage_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers.mean() * 100

percentage_outliers(car_2['odometer'])
```

```python
percentage_outliers(car_2['mmr'])
```

```python
# replacing outliers with the median
def replace_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # replacing the outliers with the median
    data = np.where((data < lower_bound) | (data > upper_bound), data.median(), data)
    return data

replace_outliers(car_2['odometer'])
replace_outliers(car_2['mmr'])
```

```python
car_2['odometer'] = replace_outliers(car_2['odometer'])
car_2['mmr'] = replace_outliers(car_2['mmr'])
```

```python
# dropping the sales_time colmn
car_3 = car_2.drop(columns=['sales_time'])
car_3.head()
```

```python
# seperating categorical and numerical features
categorical = car_3.select_dtypes(include=['object'])
numerical = car_3.select_dtypes(include=['float64'])
```

```python
categorical.head()
```

```python
# Encoding the categorical features with OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()

categorical = pd.DataFrame(encoder.fit_transform(categorical), columns=categorical.columns)
categorical.head()
```

```python
numerical.head()
```

```python
# excluding selling_price from the numerical features
target = numerical['sellingprice']
numerical = numerical.drop(columns=['sellingprice'])
```

```python
# sclain the numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numerical = pd.DataFrame(scaler.fit_transform(numerical), columns=numerical.columns)
```

```python
numerical.head()
```

```python
numerical.hist(figsize=(10, 6))
```

```python
# combining the numerical and categorical features
car_4 = pd.concat([numerical, categorical], axis=1)
car_4.head()
```

## Modelling

```python
# splitting the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(car_4, target, test_size=0.2, random_state=42)
```

```python
# shape of the training and testing sets
X_train.shape, X_test.shape
```

```python
# importing models
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# linear regression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)

# metrics
lr_mse = mean_squared_error(y_test, lr_pred)
lr_mae = mean_absolute_error(y_test, lr_pred)
lr_r2 = r2_score(y_test, lr_pred)

# printing
print('Linear Regression')
print('Mean Squared Error:', lr_mse)
print('Mean Absolute Error:', lr_mae)
print('R2 Score:', lr_r2)
```

```python
# decision tree
dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)

# metrics
dt_mse = mean_squared_error(y_test, dt_pred)
dt_mae = mean_absolute_error(y_test, dt_pred)
dt_r2 = r2_score(y_test, dt_pred)

# printing
print('Decision Tree')
print('Mean Squared Error:', dt_mse)
print('Mean Absolute Error:', dt_mae)
print('R2 Score:', dt_r2)
```
